```python
def plot_log_transforms(df, col):
    """
    need to do log + c where c is a constant. otherwise you end up with negative values after log transform. so you need to add a constant to get it over equaling zero. the mind
    # of this column is -2 so c needs to be > 2
    """
    fig, ax = plt.subplots(3,2, figsize = (16,8))
    sns.distplot(df[col], ax = ax[0,0])
    ax[0,0].set_title('Feature As is')
    # need to do log + c where c is a constant. otherwise you end up with negative values after log transform. so you need to add a constant to get it over equaling zero. the mind
    # of this column is -2 so c needs to be > 2
    sns.distplot(np.log(df[col] + abs(df[col].min())+0.01), ax = ax[0,1])
    ax[0,1].set_title('log(Feature.min()+0.01)')
    sns.distplot(np.log(df[col] + abs(df[col].min())+ 1), ax = ax[1,0])
    ax[1,0].set_title('log(Feature.min()+1)')
    sns.distplot(np.log(df[col] + abs(df[col].min()) + 5), ax = ax[1,1])
    ax[1,1].set_title('log(Feature.min()+5)')
    sns.distplot(np.log(df[col] + abs(df[col].min()) + 20), ax = ax[2,0])
    ax[2,0].set_title('log(Feature.min()+20)')
    sns.distplot(np.log(df[col]+abs(df[col].min()) + 100), ax = ax[2,1])
    ax[2,1].set_title('log(Feature.min()+100)')
    plt.tight_layout()
```

### Correlation between numeric variables?

There are several features in hhold_b_train and hhold_c_train that have a high correlation

This means an opportunity to reduce the feature set through transformations like PCA

```python
def corr(df, threshold):
    """
    Go through df and get correlations between the numeric columns
    Pass in the numeric data frame. 
    Could rewrite the function so that it will call the other func to get numeric columns
    """
    df_corr = df.corr()
    cols = df.columns
    threshold = threshold
    
    corr_list = []
    for i in range(len(df.columns)):
        for j in range(i+1, (len(df.columns) )):
            if (df_corr.iloc[i,j] >= threshold and df_corr.iloc[i,j] < 1) or (df_corr.iloc[i,j]<0 and df_corr.iloc[i,j] <= -threshold):
                corr_list.append([df_corr.iloc[i,j],i,j])
    
    s_corr_list = sorted(corr_list, key=lambda x: -abs(x[0]))            
    if s_corr_list == []:
        print('No pair of features have a correlation above threshold\n')
    else:
        for v,i,j in s_corr_list:
            print(cols[i], 'and', cols[j], '%.2f' %v)
```

### Build some models

Let's build a simple model with the hhold data.

Since each dataset is from a different country, we will train 3 seperate models, then combine the predictions to a single submission.

First convert the categorical data to onehot encoding and standardize the numeric columns

```python
def standardize(df):
    numeric = df.select_dtypes(include=['int64', 'float64'])
    
    # subtract mean and divide by std
    df[numeric.columns] = (numeric- numeric.mean()) / numeric.std()
    
    return df

def pre_process_data(df, enforce_cols = None):
    """
    Standardize the numeric columns and one hot encode the categorical columns
    """
    
    print('Input shape:\t{}'.format(df.shape))
    
    df = standardize(df)
    print('After standardization {}'.format(df.shape))
    
    # get dummies for categoricals
    df = pd.get_dummies(df)
    print('After converting categoricals:\t{}'.format(df.shape))
    df.info()
    
    # match test set and training set columns
    if enforce_cols is not None:
        to_drop = np.setdiff1d(df.columns, enforce_cols)
        to_add = np.setdiff1d(enforce_cols, df.columns)

        df.drop(to_drop, axis=1, inplace=True)
        df = df.assign(**{c: 0 for c in to_add})
    
    
    # just fill NaN with 0 for now. Later try fill it with mean or explore more to figure out how to best impute
    df.fillna(0, inplace = True) 
    
    
    return df    
```

```python
import lightgbm as lgb
lgb_params = {'objective': 'binary', 
          'learning_rate': '0.1',
          'num_threads':'4'
          }

def train_lgb(X,y):
    print('Training lgb model...')
    dstrain = lgb.Dataset(X, label = y)
#     model = lgb.train(params, dstrain, valid_sets = dstrain,
#                       num_boost_round = 8000, verbose_eval = True)
    model = lgb.LGBMClassifier(n_estimators = 10,
                              objective = 'binary', 
                              num_threads = 4,
                              learning_rate = 0.05)
    model.fit(X,y)
    
    print('Done training!')
    return model
```

```python
from sklearn.ensemble import RandomForestClassifier

def train_RF_model(features, labels, **kwargs):
    
    # instantiate model
    model = RandomForestClassifier(n_estimators=50, random_state=0)
    
    # train model
    model.fit(features, labels)
    
    # get a (not-very-useful) sense of performance
    accuracy = model.score(features, labels)
    print(f"In-sample accuracy: {accuracy:0.2%}")
    
    return model
```

```python
def make_country_sub(preds, test_feat, country):

    
    # just get probabilities for p = 1 with preds[:,1]
    country_sub = pd.DataFrame(data = preds[:,1],
                              columns = ['poor'],
                              index = test_feat.index)
    
    # add country code to join later on
    country_sub['country'] = country
    return country_sub[['country', 'poor']]
```

```python
# lgbm subs
a_sub = make_country_sub(a_lgb_preds, hhold_a_test, 'A')
b_sub = make_country_sub(b_lgb_preds, hhold_b_test, 'B')
c_sub = make_country_sub(c_lgb_preds, hhold_c_test, 'C')
sub = pd.concat([a_sub, b_sub, c_sub])
sub.to_csv('lgb_baseline.csv')
```
